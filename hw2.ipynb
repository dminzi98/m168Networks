```swift
%include "EnableIPythonDisplay.swift"
IPythonDisplay.shell.enable_matplotlib("inline")
import Python
import Foundation
```

```swift
let nx = Python.import("networkx")
let plt = Python.import("matplotlib.pyplot")
```

```swift
// network setup 
var net = nx.Graph()
let edges = [(1, 2), (1, 5), (1,6), (2, 3), (2, 5), (3, 5), (3, 4), (4, 5), (4, 7), (5, 6), (5, 7), (6, 7), (6, 8), (7, 8), (8, 9), (9, 10)]

for edge in edges {
  net.add_edge(edge.0 - 1, edge.1 - 1)
}

// preliminary check
print(net.number_of_edges())
print(net.number_of_nodes())
```

```swift
// visualization of the network
var labels: PythonObject = [1: "A", 2: "B", 3: "C", 4: "D", 5: "E", 6: "F", 7: "G", 8: "H", 9: "I", 10: "J"]
nx.draw_spring(net)
plt.show()
```

```swift
var dCent = nx.degree_centrality(net)
var cCent = nx.closeness_centrality(net)
var bCent = nx.betweenness_centrality(net)

print("Node & Degree & Closeness & Betweeness \\\\")
for node in 0...9 {
  let l: String = String(format: "%d & %.3f & %.3f & %.3f \\\\", node, Float(dCent[node])!, Float(cCent[node])!, Float(bCent[node])!)
  print(l)
}
```

```swift
var fiveReg = nx.random_regular_graph(4,8)
nx.draw_spring(fiveReg)
plt.show()
var closeC = nx.betweenness_centrality(fiveReg)
print(closeC)
```
