```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
log_n = []
shortest_paths = []

for n in range(20, 100):
    G = nx.watts_strogatz_graph(n, 20, .15)
    shortest_path = nx.average_shortest_path_length(G)
    log_n.append(np.log(n))
    shortest_paths.append(shortest_path)
```

```python
fig, ax = plt.subplots()
ax.plot(log_n, label='log(n)')
ax.plot(shortest_paths, label='shortest paths')
leg = ax.legend()
plt.ylabel('n')
plt.title('p = .15')
plt.legend()
plt.savefig('p15n100k20.png')
```

```python
p_vals = np.arange(.1, 1, 0.001)
shortest_paths = []
shortest_paths_sum = []


for i in range(10):
    for p in p_vals:
        G = nx.watts_strogatz_graph(50, 7, p)
        shortest_path = nx.average_shortest_path_length(G)
        shortest_paths.append(shortest_path)
    shortest_paths_sum += shortest_paths
    
    
```

```python
fig, ax = plt.subplots()
plt.plot(p_vals, shortest_paths)
plt.ylabel('shortest path')
plt.xlabel('p')
plt.title('Shortest Path for Probability; n=50, k=7')
plt.savefig('SPvsPR_n50.png')
```
